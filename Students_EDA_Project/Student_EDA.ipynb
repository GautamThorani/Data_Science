```python
import matplotlib.pyplot as plt
```

```python
numerical_variables = df.select_dtypes(exclude='object').columns
```

```python
for column in numerical_variables:
    plt.figure(figsize=(8, 6))
    df[column].plot(kind='hist', bins=20, edgecolor='black')
    plt.title(column)
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
```

```python
num_cols = 2
num_rows = (len(numerical_variables) + 1) // num_cols
fig, axes = plt.subplots(num_rows, num_cols, figsize=(12, 8))

for i, column in enumerate(numerical_variables):
    row = i // num_cols
    col = i % num_cols
    axes[row, col].hist(df[column], bins=20, edgecolor='black')
    axes[row, col].set_title(column)
    axes[row, col].set_xlabel('Value')
    axes[row, col].set_ylabel('Frequency')
    axes[row, col].grid(True)


for i in range(len(numerical_variables), num_rows*num_cols):
    row = i // num_cols
    col = i % num_cols
    axes[row, col].axis('off')

plt.tight_layout()
plt.show()
```
